**Observation**

The dataset contains both male and female students.
The number of female students is slightly higher than male students.

```python
sns.countplot(x='sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

**Observation:**

Most students are between 15 and 18 years old.
Very few students are older than 20.

```python
sns.histplot(df['age'], bins=10)
plt.title("Age Distribution")
plt.show()
```

**Observation**

Students with higher study time generally achieve better final grades.
Increased study time is associated with improved academic performance.

```python
sns.boxplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.show()
```

**Observation**

Students with a high number of absences tend to have lower grades.
Higher attendance is generally linked to better performance.

```python
sns.scatterplot(x='absences', y='G3', data=df)
plt.title("Absences vs Final Grade")
plt.show()
```

**Observation**

G1 and G2 have a strong positive correlation with G3.
Failures have a negative correlation with final grades.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

**Hypothesis 1**

H₀: Study time has no effect on final grades.

Result: Students with higher study time generally obtain better grades.

Conclusion: The data does not support H₀.

**Hypothesis 2

H₀: Absences do not affect final grades.

Conclusion: The data does not support H₀.

Hypothesis 3
